## Level 1

## Level 3

```python
mat = scipy.io.loadmat('LineMNK_Data2.mat')
# print(mat)
Y=mat['Y'][0]
X=mat['X'][0]
N3 = len(X)
for x,y in zip(X,Y):
    plt.scatter(x,y)
print(N3)

def our_min(f, x0, step, tol):
    xt = x0
    prevf = f(xt)
    xt+=step
    tempf = f(xt)
    while 1>0:
        if tempf > prevf:
            step*=-0.25
        prevf = tempf
        xt+=step
        tempf = f(xt)
        if (f(xt-step)>tempf and f(xt+step)>tempf):
            return xt
   
print(our_min(lambda x: x*x, 10, 0.01, 0.0001))

def our_TwoVarsMinimum(thisf, x0, step, tol, niter=10000):
    fnow = 0
    fpre = 2 * tol
    xtemp=x0[0]
    ytemp=x0[1]
    while abs(fnow-fpre)>tol:
        niter-=1
        if niter==0:
            return 'Ошибка поиска. Сработало ограничение по времени поиска'
        fpre = thisf(xtemp, ytemp)
        xtemp= our_min(lambda x: thisf(x, ytemp),xtemp,step,tol)
        ytemp= our_min(lambda y: thisf(xtemp, y),ytemp,step,tol)
        fnow = thisf(xtemp, ytemp)
    return [xtemp,ytemp]

print(our_TwoVarsMinimum(lambda x,y: x*x*y*y,[1,1],0.1,0.0001))

f = lambda x,k,b:x*k+b   
def ff(k,b):
    sum=0
    for i in range(len(X)):
        sum+=(f(X[i],k,b)-Y[i])**2
    return sum
k,b = our_TwoVarsMinimum(ff, [0,0],0.1,0.0001)
print(k,b)
print(ff(k,b))

print('y = '+str(k)+' * x + '+str(b))
```

```python
Yt=[]
for i in X:
    Yt.append(f(i,k,b))
plt.plot(X,Yt,linewidth=2,color='yellow')
for x,y in zip(X,Y):
    plt.scatter(x,y,color=(0.1,0,0.6,0.1))
```

## Level 4

```python
mat = scipy.io.loadmat('QuadMNK_Data2.mat')
# print(mat)
Y2=mat['Y'][0]
X2=mat['X'][0]
N4 = len(X2)
for x,y in zip(X2,Y2):
    plt.scatter(x,y)
    
print(N4)

fq = lambda x,a,b,c:x*x*a+b*x+c

def ffq(a,b,c):
    sum=0
    for i in range(len(X)):
        sum+=(fq(X2[i],a,b,c)-Y2[i])**2
    return sum

def our_ThreeVarsMinimum(thisf, x0, step, tol, niter=10000):
    fnow = 0
    fpre = 2 * tol
    xtemp=x0[0]
    ytemp=x0[1]
    ztemp=x0[2]
    while abs(fnow-fpre)>tol:
        niter-=1
        if niter==0:
            return 'Ошибка поиска. Сработало ограничение по времени поиска'
        fpre = thisf(xtemp, ytemp, ztemp)
        xtemp= our_min(lambda x: thisf(x, ytemp, ztemp),xtemp,step,tol)
        ytemp= our_min(lambda y: thisf(xtemp, y, ztemp),ytemp,step,tol)
        ztemp= our_min(lambda z: thisf(xtemp, ytemp, z),ztemp,step,tol)
        fnow = thisf(xtemp, ytemp, ztemp)
    return [xtemp,ytemp,ztemp]


print(our_ThreeVarsMinimum(lambda x,y,z: x*x*y*y*z*z,[1,1,1],0.1,0.0001))

a,b,c = our_ThreeVarsMinimum(ffq, [0,0,0],0.1,0.0001)
print(a,b,c)
print(ffq(a,b,c))
```

```python
Yt2=[]
for i in X2:
    Yt2.append(fq(i,a,b,c))
plt.plot(X2,Yt2,linewidth=2,color='yellow')
for x,y in zip(X2,Y2):
    plt.scatter(x,y,color=(0.1,0,0.6,0.1))
```
