```python
import random as rd
import numpy as np
import matplotlib.pyplot as plt
import scipy.io
```

```python
N=30
D=np.zeros(5000)
D=[a+rd.uniform(-5,0) for a in D]  #равномерное распределение - rd.uniform
Dmin=np.min(D)
Dmax=np.max(D)
Dmean=np.mean(D)
Hpoint=np.linspace(Dmin,Dmax,N+1)
Count=[]

plt.figure(figsize=(17,8))         # косметика графика
for i in range(1,N+1):
    CountNow=np.count_nonzero(D<Hpoint[i])-np.count_nonzero(D<Hpoint[i-1])
    Count.append(CountNow)
    plt.fill([Hpoint[i-1],Hpoint[i-1],Hpoint[i],Hpoint[i],Hpoint[i-1]],[0,CountNow,CountNow,0,0])
plt.xlabel('x')
plt.ylabel('Количество')
plt.xlim([Dmin-0.2*(Dmax-Dmin),Dmax+0.2*(Dmax-Dmin)])
plt.title('Гистограмма')
plt.grid()
plt.legend(["f(x)"])
plt.show()
```

```python
import sympy as sp
N=30
X=np.linspace(0,10,1000)
k=1
b=-5
Y=np.array([k*x+b+np.random.normal(0, 1.5) for x in X])

Dmin=np.min(X)
Dmax=np.max(X)
Sx= X.sum()
Sy= Y.sum()
X2=X*X
XY=X*Y
SX2=X2.sum()
SXY=XY.sum()
print(Sx,Sy,SX2,SXY)
K,B=sp.symbols('K,B')
f1=sp.Eq(2*K*Sx+2*B*len(X)-2*Sy,0)  #! После загрузки данных тут подстава
f2=sp.Eq(2*K*SX2+2*B*Sx-2*SXY,0)
KB=sp.solve([f1,f2],[K,B])
print(KB)
YClear=[KB[K]*x+KB[B] for x in X]
print(type(X),type(YClear))
plt.figure(figsize=(17,8))         # косметика графика
for x,y in zip(X,Y):
    plt.scatter(x,y)
#for i in range(1,N+1):
#    CountNow=np.count_nonzero(D<Hpoint[i])-np.count_nonzero(D<Hpoint[i-1])
#    Count.append(CountNow)
#    plt.fill([Hpoint[i-1],Hpoint[i-1],Hpoint[i],Hpoint[i],Hpoint[i-1]],[0,CountNow,CountNow,0,0])
plt.plot(X,YClear,color='red')
plt.xlabel('x')
plt.ylabel('Количество')
plt.xlim([Dmin-0.2*(Dmax-Dmin),Dmax+0.2*(Dmax-Dmin)])
plt.title('Гистограмма')
plt.grid()
plt.legend(["f(x)"])
plt.show()
```
